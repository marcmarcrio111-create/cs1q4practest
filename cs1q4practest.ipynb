```python
name = input("Enter the student name:")
quiz_1_score = float(input("Enter the quiz 1 score:"))
quiz_2_score = float(input("Enter the quiz 2 score:"))
formative_assessment_1_score = float(input("Enter the formative assessment 1 score:"))
formative_assessment_2_score = float(input("Enter the formative assessment 2 score:"))
final_exam_score = float(input("Enter the final exam score:"))

print("\n----- Results -----")
print(f"Hello!")

if quiz_1_score >= 35:
  print("Quiz 1 passed!")
else:
  print("Quiz 1 failed.")
if quiz_2_score >= 15:
  print("Quiz 2 passed!")
else:
  print("Quiz 2 failed.")
if formative_assessment_1_score >= 18:
  print("Formative assessment 1 passed!")
else:
  print("Formative Assessment 1 failed.")
if formative_assessment_2_score >= 24:
  print("Formative assessment 2 passed!")
else:
  print("Formative Assessment 2 failed.")
if final_exam_score >= 20:
  print("Final exam passed!")
else:
  print("Final exam failed.")
```
